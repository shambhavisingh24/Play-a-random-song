```python
def get_song_link(year, df):
    # Filter the DataFrame for the given year
    songs_from_year = df[df['year'] == year]
    
    if songs_from_year.empty:
        return "No songs found for this year."
    
    # Select a random song link
    random_song_link = random.choice(songs_from_year['fin'].tolist())
    return random_song_link
```
